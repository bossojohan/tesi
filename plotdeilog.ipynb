```python
import matplotlib.pyplot as plt
import pylab as pl
import numpy as np
epochs =[]
acc=[]
loss=[]
val_acc=[]
val_loss=[]
with open("logs/I_Ex-training-1550227970.6757874.log", "r") as in_file:        
        for line in in_file:             
            data = line.split(',')
            epochs.append(data[0])
            acc.append(data[1])
            loss.append(data[2])
            val_acc.append(data[3])
            val_loss.append(data[4])

    
nacc= np.array(acc[1:],dtype='float32')   
nvacc= np.array(val_acc[1:],dtype='float32') 
nloss= np.array(loss[1:],dtype='float32')     
nvloss= np.array(val_loss[1:],dtype='float32')
nepo = np.array(epochs[1:])


plt.plot(nepo,nacc,'bo', label= 'train accuracy')
plt.plot(nepo,nvacc,'b', label= 'validation accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend()
plt.show()            
```
